```python
# Kevin King
# CS 74 HW 1 Q3
# Spring 2022

import matplotlib.pyplot as plt
%matplotlib inline 

# gradient descent function
# inputs: alpha (learning rate parameter), max_its (maximum number of iterations), w0 (initialization)
def gradient_descent(alpha,max_its,w0):
    ##Your code here
    cost_history = []
    
    for k in range(0, max_its+1):
        gw = (1/50)*(w0**4 + w0**2 + 10*w0 - 50)
        dgdw = (1/50)*(4*w0**3 + 2*w0 + 10)
        w0 = w0 - alpha*dgdw
        cost_history.append(gw)
    
    return cost_history

first_run = gradient_descent(1,1000,2.0)
second_run = gradient_descent(0.1,1000,2.0)
third_run = gradient_descent(0.01,1000,2.0)

# create x axis
x = [None]*1001
for k in range(0, 1001):
    x[k] = k

# include legend?
plt.plot(x, first_run, 'r--')
plt.plot(x, second_run, 'b--')
plt.plot(x, third_run, 'g--')
plt.xlabel('Step k')
plt.ylabel('Cost')
```
